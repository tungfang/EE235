## Assignment 1 -- Matched Filter with Ideal Signals

```python
# Assignment 1 - Matched Filter with Ideal Signals 

# Part A
fs = 1000
th = np.arange(0, 1, 1/fs)
u = np.heaviside(th, 1)
s1 = u - np.heaviside(th-1, 1)
s0 = -s1

# Part B
y1 = np.convolve(s1, s1)/fs
y0 = np.convolve(s1, s0)/fs

# Part C
t_y = np.arange(0, len(y1), 1)/fs

fig1 = plt.figure(1)
fig1.subplots_adjust(hspace = 2, wspace = 1)

plt.subplot(2, 1, 1)
plt.title('y1 vs. t_y')
plt.plot(t_y, y1, color = 'r')
plt.ylabel('Amplitude')
plt.xlabel('Time')

plt.subplot(2, 1, 2)
plt.title('y0 vs. t_y')
plt.plot(t_y, y0, color = 'g')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.show()

# Part D
y1_corr = y1[len(s1)]
y0_corr = y0[len(s0)]
print(y1_corr, y0_corr)
```

## Assignment 2 -- Signal Decoding with a Matched Filter 

```python
# Assignment 2 -- Signal Decoding with a Matched Filter 

# Part A
# Decode function take inputs signal r, the sampling rate fs, 
# the 0-signal s0, and the 1-signal s1
# and return the decoded symbol, which is either 0 or 1.

def decode(r, fs, s0, s1):
    y0 = np.convolve(r, s0) / fs
    y1 = np.convolve(r, s1) / fs
    y0_corr = y0[len(s1)]
    y1_corr = y1[len(s1)]
    if (y1_corr > y0_corr):
        s = '1'
    else:
        s = '0'
    return s
```

```python
# Part B
n = np.random.normal(0,3,len(s1))
r = s1 + n
r0 = s0 + n
sc = np.concatenate([s1, s0])
rc = np.concatenate([r, r0])
t = np.arange(0, len(rc), 1) /fs
fig2 = plt.figure(2)
fig2.subplots_adjust(hspace = 2, wspace = 1)

plt.subplot(1, 1, 1)
plt.title('signals with and without noise')
plt.plot(t, rc, label = 'with noise', color = 'r')
plt.plot(t, sc, label = 'without noise', color = 'b')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.legend()
plt.show()

# Part C
yr1 = np.convolve(r, s1)/fs
yr0 = np.convolve(r, s0)/fs
t_yr = np.arange(0, len(yr0), 1) / fs

fig3 = plt.figure(3)
fig3.subplots_adjust(hspace = 2, wspace = 1)

plt.subplot(1, 1, 1)
plt.title('yr1(t) and yr0(t)')
plt.plot(t_yr, yr1, label = 'yr1', color = 'r')
plt.plot(t_yr, yr0, label = 'yr0', color = 'b')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.legend()
plt.show()

yr1_corr = yr1[len(s1)]
yr0_corr = yr0[len(s0)]
print(yr1_corr, yr0_corr)

# Part D
s1n = decode(s1 + n, fs, s0, s1)
s0n = decode(s0 + n, fs, s0, s1)
print(s1n, s0n)
```

```python
# Assignment 3 -- Decipher Recived Message
# part A
fs = 8000
t = np.arange(0, 1, 1/fs)
u = np.heaviside(t, 1)
s1 = u - np.heaviside(t-1, 1)
s0 = -s1

file = open('receivedmsg.csv', 'rb')
rm = np.loadtxt(file, delimiter=',')
array = np.arange(0,rm.shape[0])
for i in range (0, rm.shape[0]):
    k = decode(rm[i], fs, s0, s1)
    array[i] = k
print(array)
```
