## Assignment 2: Amplitude Operations on Signals

```python
# Assignment 2 - Amplitude Operations on Signals

# t = numpy.arange(start, stop, ts) # ts = sampling period (step size)

# Part A
fs = 2
t = np.arange(0, 3.5, 1/fs)
x = t * 0.5
y = t * t 

# Part B
z = x - 2 * y

# Part C
nt = int(2*fs)
nstart = int(0*fs)
w1 = z[nt-nstart]
print(w1) 

# Part D
n2 = int(1.5*fs)
w2 = z[nstart : nstart - n2]
print(w2)
```

### Discussion
fs = 1 has a lower frequency and means a smaller step size and less samplings are kept track of. Because the frequency is cut by half, it results the length in x, y, and z also decrease by half. In addition, the length of w2 will shrink by half to 2 instead of 4. However, the result of w1 stays the same since we're only taking one value

## Assignment 5 -- Plotting Sound Files in Subplots

```python
# Assignment 5 -- Plotting Sound Files using Subplots

# Part A
from scipy.io import wavfile as wav
fs1, y1 = wav.read('train32.wav')
fs2, y2 = wav.read('tuba11.wav')
t1 = np.arange(0, len(y1), 1)
t1 = t1 / fs1 * 1000
t2 = np.arange(0, len(y2), 1)
t2 = t2 / fs2 * 1000

# Part B
fig2 = plt.figure(2)
fig2.subplots_adjust(hspace = 0.5, wspace = 0.7)

plt.subplot(1, 2, 1)
plt.plot(t1, y1, color = 'purple')
plt.title('y1')
plt.ylim(-30000, 30000)
plt.xlim(0, 2500)
plt.ylabel('Amplitude')
plt.xlabel('Time(ms)')

plt.subplot(1, 2, 2)
plt.plot(t2, y2, color = 'gold')
plt.title('y2')
plt.ylim(-30000, 30000)
plt.xlim(0, 2500)
plt.ylabel('Amplitude')
plt.xlabel('Time(ms)')
plt.show()
```
